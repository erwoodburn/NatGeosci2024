```python
# --- Utility Functions ---

def read_nldas(wy, time, var):
    #f  = '../baseline/nldas/wy{}/NLDAS.{}.{:06}.pfb'.format(wy, var, time)
    f  = '/global/cfs/projectdirs/ecoslim/baseline/nldas/wy{}/NLDAS.{}.{:06}.pfb'.format(wy, var, time)
    dd = pf.read_pfb(f)[0,:,:] # 2d surface array
    return dd

def read_clm(_case, wy, time):
    '''3D array variables:
       0.eflx_lh_tot, 1.eflx_lwrad_out, 2.eflx_sh_tot, 3.eflx_soil_grnd, 4.qflx_evap_tot, 5.qflx_evap_grnd, 6.qflx_evap_soi,
       7.qflx_evap_veg, 8.qflx_tran_veg, 9.qflx_infl, 10.swe_out, 11.t_grnd, 12:end.t_soil layers'''  
    if _case == 'baseline':
        #f  = '../{}/wy{}/bl_wy{}.out.clm_output.{:05}.C.pfb'.format(_case,wy,wy,time)
        f  = '/global/cfs/projectdirs/ecoslim/{}/{}/wy{}/bl_wy{}.out.clm_output.{:05}.C.pfb'.format(_case,_case,wy,wy,time)
    else:
        #f  = '../{}/wy{}/{}_wy{}.out.clm_output.{:05}.C.pfb'.format(_case,wy,_case,wy,time)
        f  = '/global/cfs/projectdirs/ecoslim/{}/{}/wy{}/{}_wy{}.out.clm_output.{:05}.C.pfb'.format(_case,_case,wy,_case,wy,time)
    dd = pf.read_pfb(f)
    return dd 

def find_times(wy):
    #fnames = glob.glob('../baseline/nldas/wy{}/*APCP*'.format(wy))
    fnames = glob.glob('/global/cfs/projectdirs/ecoslim/baseline/nldas/wy{}/*APCP*'.format(wy))
    times  = []
    for f in fnames:
        times.append(int(f.split('.')[-2]))
    times = np.array(times)
    return np.sort(times)

def wtd_percentile(data, weights, perc=[0.1,0.5,0.9]):
    #perc : percentile in [0-1]
    ix = np.argsort(data)
    data = data[ix] # sort data
    weights = weights[ix] # sort weights
    cdf = (np.cumsum(weights) - 0.5 * weights) / np.sum(weights) # 'like' a CDF function
    try:
        return np.append(np.interp(perc, cdf, data), [np.average(data, weights=weights), weights.sum()])
    except ValueError:
        return (len(perc)+2)*[np.nan]
```

```python
# --- Read in all ParFlow Pressure fields ---

class build_pf_Q():
    def __init__(self, wy_list):
        
        self.wy_list = wy_list
        
        self.pf_Q_dict       = {}
        self.surf_press_dict = {}
        self.mod_dates_dict  = {}
        
    def pull_press_files(self, wy): 
        # ParFlow pressure files
        _dir    = 'wy{}'.format(wy)     # directory
        if _case == 'baseline':
            _prefix = 'bl_wy{}'.format(wy)  # parflow output prefix
        else:
            _prefix = '{}_wy{}'.format(_case, wy)
        var = 'press'
        ff = []
        
        for file in os.listdir('/global/cfs/projectdirs/ecoslim/{}/{}/{}'.format(_case,_case,_dir)):
            if file.endswith('pfb') & file.startswith(_prefix):
                if all(x in file.split('.') for x in [var]):
                    ff.append(os.path.join('/global/cfs/projectdirs/ecoslim/{}/{}/{}'.format(_case,_case,_dir), file))
        ff.sort()
        return ff


    def loop_files(self, wy):
        # Read in Parflow Q and surface pressure 
        #pdb.set_trace()
        ff = self.pull_press_files(wy)

        _pf_Q  = np.zeros([len(ff)])  # parflow overland flow at PumpHouse [m3/hr]
        _surf_press = np.zeros([len(ff), mask.shape[0], mask.shape[1]])  # gridded surface pressure [m]

        for i,f in enumerate(ff):
            _pressure = pf.read_pfb(f)
            _surf_press[i] = _pressure[-1]   

            _Q_grid = pftools.hydrology.calculate_overland_flow_grid(pressure=_pressure, slopex=slopex, slopey=slopey, mannings=mannings, dx=dx, dy=dy, flow_method='OverlandFlow', mask=mask_)
            _pf_Q[i] = _Q_grid.max()

        _mod_dates = pd.date_range('{}-10-01 00:00:00'.format(wy-1), freq='1H', periods=len(ff))

        self.pf_Q_dict[wy]       = _pf_Q
        self.surf_press_dict[wy] = _surf_press
        self.mod_dates_dict[wy]  = _mod_dates  
        
        
    def runit(self):
        for wy in self.wy_list:
            print ('Working on ', wy)
            self.loop_files(wy)
```

```python
# --- Overland Flow Fractions ---

# --- Elevation bands for binning ---
zmin, zmax = dem[mask>0].min(), dem[mask>0].max()
zbands     = np.linspace(zmin-10, zmax+10, 8)
zbands_arr = np.column_stack([zbands[:-1],zbands[1:]])
zbands_mu  = zbands_arr.mean(axis=1)

zin = dem[mask>0].ravel()
zbandsin = (zbands_arr[:,0][:, np.newaxis] <= zin) & (zin < zbands_arr[:,1][:, np.newaxis]) # 2D array with shape like zbands_arr


class build_over_frac():
    def __init__(self, wy_list):
        self.wy_list = wy_list
        
        # --- Dictionaries to save outputs ---
        self.overland_rain_part_dict = {}
        self.overland_snow_part_dict = {}

        self.p_basin_dict   = {}
        self.swe_basin_dict = {}
        
        # --- Previous spatial swe field ---
        self.swe_t0_glob = np.zeros_like(mask_[-1]) # last swe value, gets overwritten at end of water year 

    # --- Loop over times and calculate overland flow Rain/Snow and elevation partitioning ---
    def loop_files(self, wy):
        _times  = find_times(wy)
        
        overland_rain_part = np.zeros((len(_times)-1,len(zbands_mu)))
        overland_snow_part = np.zeros((len(_times)-1,len(zbands_mu)))

        p_basin   = []
        swe_basin = []

        #swe_t0 = np.zeros_like(mask_[-1]) # initialize swe
        swe_t0 = self.swe_t0_glob

        for i,t in enumerate(_times[1:]): # clm starts at index 1
            # Read NLDAS Precip.
            P = read_nldas(wy, t, 'APCP')*3600 + 1.e-7 # mm/s to mm/hr

            # Read CLM variables
            _clm  = read_clm(_case, wy, t)
            et    = _clm[4]*3600 # mm/s to mm/hr
            infil = _clm[9]*3600*0.1 # mm/s to mm/hr and multiply by top ParFlow layer thickness
            swe   = _clm[10] # clm is mm, but at hourly timesteps dSWE should be mm/hr

            p_basin.append(np.median(P[mask>0]))
            swe_basin.append(np.percentile(swe[mask>0],[5,50,95]))

            # Calculate overland flow as P-ET-inf
            # but, use delta SWE rather than P if there is a melting snowpack
            dswe = swe_t0-swe # mm/hr - negative is gaining snowpack, positive is melting snowpack, zero is no change

            #influx = np.where((swe==0.0)&(dswe==0.0), P, dswe) # if no swe and dswe=0, take P; if swe, take dswe
            influx = np.where(swe<10, P, dswe) # if no swe, take P; if swe, take dswe
            influx[influx<=0.0] = 1.e-7 # set dswe<0 (gaining snowpack) as influx=0.0 

            overland_flux = influx - et #- infil
            overland_flux[overland_flux<=0.0] = 1.e-7 # only interested in net positive fluxes
            overland_flux = np.where(surf_press_dict[wy][t]>0.001, overland_flux, 1.e-7) # only interested in overland flow when pressure head > 0.0

            # Spatial flux of Rain and Snow that is overland flow
            #rain_flux = np.where((swe==0.0)&(dswe==0.0), overland_flux, 0.0) 
            rain_flux = np.where(swe<10,  overland_flux, 0.0) 
            snow_flux = np.where(swe>=10, overland_flux, 0.0) 

            # Fraction of total flux from rain vs snow
            rain_frac = np.divide(rain_flux[mask>0].sum(), overland_flux[mask>0].sum(), where=overland_flux[mask>0].sum()!=0)
            snow_frac = 1 - rain_frac

            # Fraction of Rain/Snow from different elevation bands
            rain_zbands = [] # rain flux in given elevation band from zbands_arr
            for z in range(len(zbandsin)):
                rain_zbands.append(rain_flux[mask>0][zbandsin[z]].sum())
            #rain_zbands_frac = np.divide(rain_zbands, sum(rain_zbands), out=np.full_like(rain_zbands, fill_value=np.nan), where=sum(rain_zbands)!=0.0)
            rain_zbands_frac = np.divide(rain_zbands, sum(rain_zbands), out=np.zeros_like(rain_zbands), where=sum(rain_zbands)!=0.0)

            snow_zbands = [] # snow flux in given elevation band from zbands_arr
            for z in range(len(zbandsin)):
                snow_zbands.append(snow_flux[mask>0][zbandsin[z]].sum())
            #snow_zbands_frac = np.divide(snow_zbands, sum(snow_zbands), out=np.full_like(snow_zbands, fill_value=np.nan), where=sum(snow_zbands)!=0.0)
            snow_zbands_frac = np.divide(snow_zbands, sum(snow_zbands), out=np.zeros_like(snow_zbands), where=sum(snow_zbands)!=0.0)

            # Save arrays of rain/snow elevation groupings
            overland_rain_part[i] = rain_zbands_frac*rain_frac
            overland_snow_part[i] = snow_zbands_frac*snow_frac

            # Update last swe to current timestep
            swe_t0 = swe.copy() 

        self.overland_rain_part_dict[wy] = overland_rain_part
        self.overland_snow_part_dict[wy] = overland_snow_part

        self.p_basin_dict[wy]   = p_basin
        self.swe_basin_dict[wy] = swe_basin

        self.swe_t0_glob = swe_t0.copy() # Update global swe to be used for next year
        
    def runit(self):
        for wy in self.wy_list:
            print ('Working on ', wy)
            self.loop_files(wy)
```

```python
# ---------------
# --- EcoSLIM ---
# ---------------

# Loop over each WY and only store needed variables 

# 0.Time, 1.X, 2.Y, 3.Z, 4.Age, 5.Saturated_Age, 6.Mass, 7.Source (1=IC, 2=Rain, 3=Snow), 8.Exit_Status (1=outflow, 2=ET)
# 9.ParticleID, 10.InitialX, 11.InitialY, 12.InitialZ, 13.InsertTime, 14.Path_Length, 15.Saturated_Path_Length

_dir = '/global/cfs/cdirs/ecoslim/{}/post_proc/'.format(_case)


class build_eco():
    def __init__(self, wy):
        
        self.wy      = wy
  
        self.eco_raw      = None # Array of all EcoSLIM outflow particles
        self.eco_Q        = None # Array with all EcoSLIM outflow particles AFTER tau_cut and press_cut
        self.eco_df_Q_hr  = None # Dataframe with EcoSLIM particle masses aggregated to hourly
        self.pf_df_Q_hr   = None # Dataframe with ParFlow Q at pumphouse
        self.eco_src_hr   = None # Dataframe with EcoSLIM hourly Init/Rain/Snow partitioning
        self.overland_src_mass_hr  = None # Dataframe with ParFlow Overland flow Rain/Snow paritioning at different elevation bands (kg/hr)
        self.overland_particle_arr = None # Array with ParFlow overland flow particle attributes ready for convolution integral
        self.outflow_df = None # Array with all EcoSLIM particles
        self.O18_pred   = None # Predictions of 18O timeseries
        
        
    def eco_read(self):
        # This only needs to be run once per year -- does not change with parameter pertubations
        # Output from build_exit.base.py 
        _f   = os.path.join(_dir,'{}_wy{}.txt'.format(_case,self.wy))
        _eco = np.loadtxt(_f)

        # seperate Q vs ET outflow particles
        # this is all the ecoslim particle that leave as outflow without post-processing
        self.eco_raw  = _eco[np.where(_eco[:,8]==1,True,False)]
        
        
    def eco_parse(self, tau_cut, press_cut):
        # Needs to be re-run if tau_cut or press_cut change
        #pdb.set_trace()
        
        # only want outflow particles with age > 1 hr
        eco_Q = self.eco_raw[np.where(self.eco_raw[:,4]>tau_cut,True,False)]

        # Drop Q particles that exit in cells with small pressure head
        eco_Q = eco_Q[np.where(surf_press_dict[self.wy][eco_Q[:,0].astype(int), (eco_Q[:,2]/100).astype(int), (eco_Q[:,1]/100).astype(int)]>=press_cut, True, False)]

        # EcoSLIM Surface Flow output times parser
        mod_dates_Q_start = np.datetime64('{}-10-01 00:00:00'.format(wy-1))
        out_dates_Q = mod_dates_Q_start + eco_Q[:,0].astype('timedelta64[h]')
        mod_dates_Q = pd.DatetimeIndex(out_dates_Q).unique()

        out_dates_Q = pd.DatetimeIndex(out_dates_Q)
        in_dates_Q  = pd.DatetimeIndex(out_dates_Q - eco_Q[:,4].astype('timedelta64[h]'))

        # dates are with respect to particle output times
        eco_df   = pd.DataFrame(index=out_dates_Q, data=eco_Q[:,[0,1,2,4,6,7,8,10,11,14]],
                                columns=['Time','X','Y','Age','Mass','Source','Exit_Status','Xin','Yin','Path_Length'])
        # EcoSLIM Hourly Q
        self.eco_df_Q_hr = eco_df.groupby(pd.Grouper(freq='H')).sum()['Mass'] # kg/hr
        
        
        # --- EcoSLIM Rain/Snow Partitioning ---
        # this is just for plotting
        # EcoSLIM mass split at every particle
        init = np.where(eco_df['Source']==1, 1, 0)
        rain = np.where(eco_df['Source']==2, 1, 0)
        snow = np.where(eco_df['Source']==3, 1, 0)
        src_bin = np.column_stack((init,snow,rain))
        eco_src_mass = pd.DataFrame(index=eco_df.index, data=src_bin * eco_df['Mass'].to_numpy()[:,np.newaxis], columns=['init','snow','rain'])

        self.eco_src_hr  = eco_src_mass.groupby(by=pd.Grouper(freq='H')).sum()
        self.eco_Q       = eco_Q

    
    def overland_parse(self, mod_dates_dict, pf_Q_dict, overland_rain_part_dict, overland_snow_part_dict):
        # Needs to be re-run if tau_cut or press_cut change
        
        # --- Overland Flow mass arrays in kg/hr for the overland flow rain and snow comps ---
        # Parflow Hourly Q
        self.pf_df_Q_hr  = pd.DataFrame(index=mod_dates_dict[self.wy], data=pf_Q_dict[self.wy], columns=['m3h'])['m3h']
        # ! update ! do some smoothing because hourly causes issues with plotting
        #self.pf_df_Q_hr  = self.pf_df_Q_hr.rolling(window='3D').mean()
        # ! end update !
        
        # Overland Flow Flux (m3/hr) as difference between Parflow Q and EcoSLIM Q 
        # ! update ! do some smoothing because hourly causes issues
        #self.eco_df_Q_hr = self.eco_df_Q_hr.rolling(window='3D').mean()
        # ! end update !
        Q_df_over_hr = pd.DataFrame(index=self.eco_df_Q_hr.index, data=self.pf_df_Q_hr.to_numpy().ravel()[1:] - self.eco_df_Q_hr.to_numpy().ravel()/1000)
        # Truncate negative values
        Q_df_over_hr[Q_df_over_hr<0.0] = 0.0

        # Overland Flow in kg/hr (mostly for plotting)
        overland_rain_mass   = pd.DataFrame(index=Q_df_over_hr.index, data=1000*overland_rain_part_dict[self.wy]*Q_df_over_hr.to_numpy().ravel()[:,np.newaxis], columns=zbands_mu.astype(int))
        overland_snow_mass   = pd.DataFrame(index=Q_df_over_hr.index, data=1000*overland_snow_part_dict[self.wy]*Q_df_over_hr.to_numpy().ravel()[:,np.newaxis], columns=zbands_mu.astype(int))
        self.overland_src_mass_hr = pd.DataFrame(index=overland_rain_mass.index, data=np.column_stack((overland_rain_mass.sum(axis=1), overland_snow_mass.sum(axis=1))), columns=['rain','snow'])

        
        # --- Make arrays to match the EcoSLIM format ---
        _mass_rain  = overland_rain_mass.to_numpy().ravel()
        _times_rain = np.array([[i]*len(zbands_mu) for i in np.arange(1, len(overland_rain_mass.index)+1)]).ravel() # output model times
        _src_rain   = np.ones_like(_times_rain)*2    # particle input source (rain=2, snow=3)
        _tau_rain   = np.ones_like(_times_rain)*0.1  # particle residence time (hours) that I am making up
        _zin_rain   = np.array(overland_rain_mass.columns.to_list()*(len(overland_rain_mass.index))) # particle input elevatoin

        _mass_snow  = overland_snow_mass.to_numpy().ravel()
        _times_snow = np.array([[i]*len(zbands_mu) for i in np.arange(1, len(overland_snow_mass.index)+1)]).ravel() # output model times
        _src_snow   = np.ones_like(_times_snow)*3    # particle input source (rain=2, snow=3)
        _tau_snow   = np.ones_like(_times_snow)*0.1  # particle residence time (hours) that I am making up
        _zin_snow   = np.array(overland_snow_mass.columns.to_list()*(len(overland_snow_mass.index))) # particle input elevatoin

        overland_particle_rain = np.column_stack((_times_rain,_tau_rain,_mass_rain,_zin_rain,_src_rain))
        overland_particle_snow = np.column_stack((_times_snow,_tau_snow,_mass_snow,_zin_snow,_src_snow))

        overland_particle_arr = np.concatenate((overland_particle_rain, overland_particle_snow))

        # sort based on model time
        self.overland_particle_arr = overland_particle_arr[overland_particle_arr[:,0].argsort()]
        
    def sim_180(self, snow_mu, rain_mu, lapse_slope, rc_elev, add_overland=False):
        # Needs to be re-run if tau_cut or press_cut change
        # Needs to be re-run if snow_mu, rain_mu, or lapse_mu change
        
        mod_dates_start = np.datetime64('{}-10-01 00:00:00'.format(self.wy-1))

        # add in overland flow components from overland_particle_arr above
        # 0:model time, 1:tau, 2:mass, 3:Zin, 4:source
        if add_overland:
            zin = dem[(self.eco_Q[:,11]/100).astype(int),(self.eco_Q[:,10]/100).astype(int)]
            _zin  = np.concatenate((zin, self.overland_particle_arr[:,3]))
            _src  = np.concatenate((self.eco_Q[:,7], self.overland_particle_arr[:,4]))
            _mass = np.concatenate((self.eco_Q[:,6], self.overland_particle_arr[:,2]))
            _time = np.concatenate((self.eco_Q[:,0], self.overland_particle_arr[:,0]))
            _age  = np.concatenate((self.eco_Q[:,4], np.zeros_like(self.overland_particle_arr[:,0]))) # assume overland flow has age=0
            
            out_dates  = mod_dates_start + _time.astype('timedelta64[h]')
            self.outflow_df = pd.DataFrame(index=out_dates, data=np.column_stack((_mass,_age,_zin,_src)), columns=['mass','age','zin','src']) 
        else:
            _zin  = dem[(self.eco_Q[:,11]/100).astype(int),(self.eco_Q[:,10]/100).astype(int)]
            _xin  = (self.eco_Q[:,11]/100).astype(int)
            _yin  = (self.eco_Q[:,10]/100).astype(int)
            _src  = self.eco_Q[:,7]
            _mass = self.eco_Q[:,6]
            _time = self.eco_Q[:,0]
            _age  = self.eco_Q[:,4]
            
            out_dates  = mod_dates_start + _time.astype('timedelta64[h]')
            self.outflow_df = pd.DataFrame(index=out_dates, data=np.column_stack((_mass,_age,_xin,_yin,_zin,_src)), columns=['mass','age','xin','yin','zin','src']) 
        
        _pred_snow_mu = np.where(_src==3, snow_mu + lapse_slope*(_zin-rc_elev)/1, 0)
        _pred_rain_mu = np.where(_src==2, rain_mu + lapse_slope*(_zin-rc_elev)/1, 0)
        _pred_ic_mu   = np.where(_src==1, (snow_mu+rain_mu)/2 + lapse_slope*(_zin-rc_elev), 0)

        _pred_mu = np.column_stack([_pred_snow_mu,_pred_rain_mu,_pred_ic_mu]).sum(axis=1)

        pred_df = pd.DataFrame(index=out_dates, data=np.column_stack((_pred_mu, _mass)), columns=['O18in','mass'])
        #pred_df['prod'] = pred_df['O18in']*pred_df['mass']
        #pred_df_sum = pred_df.groupby(by=pd.Grouper(freq='H')).sum()
        #self.O18_pred = pred_df_sum['prod'] / pred_df_sum['mass']
        
        _O18_pred = pred_df.groupby(by=pd.Grouper(freq='1H')).apply(lambda x : pd.Series(wtd_percentile(data=x['O18in'], weights=x['mass'])).T)
        _O18_pred.columns = ['10','50','90','mu','mass']
        self.O18_pred = _O18_pred
```

```python
#
tau_cut     = 0.0
#tau_cut    = 1.0

#press_cut  = 0.0
press_cut   = 0.005 
#
```

```python
# --- Loop Through Water Years ---
# Constant parameters 

# Store only what is needed for plotting

# Dictionaries to hold output
O18_sims    = {}
O18_sims_of = {}

pf_df_Q_hr  = {}
eco_df_Q_hr = {}

eco_src_hr  = {}
overland_src_mass_hr = {}

outflow_df    = {}
outflow_df_of = {}

for wy in _wys:
    _sim = build_eco(wy)
    _sim.eco_read()
    
    _sim.eco_parse(tau_cut, press_cut)
    # save some output
    eco_src_hr[wy]  = _sim.eco_src_hr
    eco_df_Q_hr[wy] = _sim.eco_df_Q_hr
    
    _sim.overland_parse(mod_dates_dict, pf_Q_dict, overland_rain_part_dict, overland_snow_part_dict)
    # save some output
    overland_src_mass_hr[wy] = _sim.overland_src_mass_hr
    pf_df_Q_hr[wy] = _sim.pf_df_Q_hr
    
    _sim.sim_180(snow_mu, rain_mu, lapse_slope, rc_elev, add_overland=False)
    # save some output
    O18_sims[wy]   = _sim.O18_pred 
    outflow_df[wy] = _sim.outflow_df
    
    _sim.sim_180(snow_mu, rain_mu, lapse_slope, rc_elev, add_overland=True)
    # save some output
    O18_sims_of[wy]   = _sim.O18_pred
    outflow_df_of[wy] = _sim.outflow_df

del (_sim)
```

```python
# EcoSLIM Age dynamics
def wtd_percentile(data, weights, perc=[0.1,0.5,0.9]):
    #perc : percentile in [0-1]
    #pdb.set_trace()
    ix = np.argsort(data)
    data = data[ix] # sort data
    weights = weights[ix] # sort weights
    cdf = (np.cumsum(weights) - 0.5 * weights) / np.sum(weights) # 'like' a CDF function
    try:
        return np.append(np.interp(perc, cdf, data), [np.average(data, weights=weights), weights.sum()])
    except ValueError:
        return (len(perc)+2)*[np.nan]


eco_age_base = outflow_df_base.copy().groupby(by=pd.Grouper(freq='1H')).apply(lambda x : pd.Series(wtd_percentile(data=x['age'], weights=x['mass'])))
eco_age_base.columns = ['10','50','90','mu','mass']

eco_age_2_5 = outflow_df_2_5.copy().groupby(by=pd.Grouper(freq='1H')).apply(lambda x : pd.Series(wtd_percentile(data=x['age'], weights=x['mass'])))
eco_age_2_5.columns = ['10','50','90','mu','mass']

eco_age_4_0 = outflow_df_4_0.copy().groupby(by=pd.Grouper(freq='1H')).apply(lambda x : pd.Series(wtd_percentile(data=x['age'], weights=x['mass'])))
eco_age_4_0.columns = ['10','50','90','mu','mass']
```
